#Smoothing and Blurring

OpenCV offers a range of functions for image smoothing and blurring, essential for reducing noise and enhancing images. These techniques are widely used in tasks such as image denoising and preprocessing. Let's explore some common smoothing and blurring techniques in OpenCV:


*   Average Blur: The average blur applies a simple averaging filter to the image, which helps to smoothen the image while reducing noise. It replaces each pixel value with the average value of its neighbouring pixels.

*  Gaussian Blur: Gaussian blur is a popular blurring technique that uses a Gaussian filter to reduce noise and preserve edges. It applies a weighted average to the pixels, giving more importance to nearby pixels. The degree of blurring is controlled by the standard deviation of the Gaussian kernel.

* Median Blur: Median blur is a non-linear filtering technique that replaces each pixel's value with the median value of its neighbouring pixels. It particularly effectively reduces salt-and-pepper noise, where isolated pixels have extreme values.

* Bilateral Blur: The bilateral blur is a powerful smoothing technique that preserves edges while reducing noise. It applies a weighted average based on spatial distance and intensity difference between pixels. This technique is handy for smoothing images while keeping essential details.


The following sections will explore these techniques in detail and provide practical examples to demonstrate their usage.

#Loading Image with OpenCV


```python
#read the image 
img = cv2.imread(r"/content/licence plate.png")

#convert the color format to RGB using imutils
img = imutils.opencv2matplotlib(img)

```

#Average blurring

```python
# Define a kernel size
kernel_size = [(3, 3), (7, 7), (9, 9), (15, 15)]

# Create a subplot grid
fig, axs = plt.subplots(1, len(kernel_size) + 1, figsize=(16, 4))

# Plot the Original Image
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Preallocate list for blurred images
blurred_images = [None] * len(kernel_size)

# Loop over the kernel sizes
for i, (kX, kY) in enumerate(kernel_size, start=1):
    # Apply an avg blur to the image
    blurred_images[i - 1] = cv2.blur(img, (kX, kY))

    # Plot the result in the corresponding subplot
    axs[i].imshow(blurred_images[i - 1])
    axs[i].set_title('Average: {}, {}'.format(kX, kY))
    axs[i].axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

#Gaussian blurring

```python
# Define a kernel size
kernel_size = [(3, 3), (7, 7), (9, 9), (15, 15)]

# Create a subplot grid
fig, axs = plt.subplots(1, len(kernel_size) + 1, figsize=(16, 4))

# Plot the Original Image
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Preallocate list for blurred images
blurred_images = [None] * len(kernel_size)

# Loop over the kernel sizes
for i, (kX, kY) in enumerate(kernel_size, start=1):
    # Apply Gaussian blur to the image
    blurred_images[i - 1] = cv2.GaussianBlur(img, (kX, kY), 0)

    # Plot the result in the corresponding subplot
    axs[i].imshow(blurred_images[i - 1])
    axs[i].set_title('Gaussian: {}x{}'.format(kX, kY))
    axs[i].axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

#Median blurring

```python
# Define a kernel size
kernel_size = [(3, 3), (7, 7), (9, 9), (15, 15)]

# Create a subplot grid
fig, axs = plt.subplots(1, len(kernel_size) + 1, figsize=(16, 4))

# Plot the Original Image
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Preallocate list for blurred images
blurred_images = [None] * len(kernel_size)

# Loop over the kernel sizes
for i, (kX, kY) in enumerate(kernel_size, start=1):
    # Apply median blur to the image
    blurred_images[i - 1] = cv2.medianBlur(img, kX)

    # Plot the result in the corresponding subplot
    axs[i].imshow(blurred_images[i - 1])
    axs[i].set_title('Median: {}x{}'.format(kX, kY))
    axs[i].axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Define bilateral filtering parameters
params = [(21, 51, 11), (31, 41, 21), (41, 61, 51)]



# Create a subplot grid
fig, axs = plt.subplots(1, len(params) + 1, figsize=(16, 4))

# Plot the Original Image
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Loop over the parameters
for i, (diameter, sigmaColor, sigmaSpace) in enumerate(params, start=1):
    # Apply bilateral filtering to the image using the current set of parameters
    blurred = cv2.bilateralFilter(img, diameter, sigmaColor, sigmaSpace)
    
    # Display the blurred image
    axs[i].imshow(blurred)
    axs[i].set_title('Blurred d={}, sc={}, ss={}'.format(diameter, sigmaColor, sigmaSpace))
    axs[i].axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
